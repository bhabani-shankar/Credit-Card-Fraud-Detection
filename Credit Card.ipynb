```python
columns = [c for c in columns if c not in ["Class"]]
```

```python
X = data.iloc[:,0:30]
y = data.iloc[:,30]
```

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
classifier = IsolationForest(max_samples=len(X),contamination='legacy',random_state=None)
```

```python
y_pred = classifier.fit_predict(X)
#scores_pred = classifier.negative_outlier_factor_
```

```python
y_pred[y_pred == 1] = 0
y_pred[y_pred == -1] = 1
print((y_pred != y).sum())
```

```python
print(accuracy_score(y, y_pred))
print(classification_report(y, y_pred))
```

```python
classifiers = LocalOutlierFactor(n_neighbors=20,contamination='auto')
```

```python
y_preds = classifiers.fit_predict(X)
```

```python
y_preds[y_preds == 1] = 0
y_preds[y_preds == -1] = 1
print(accuracy_score(y, y_preds))
```

```python
print((y_preds != y).sum())
```

```python
print(classification_report(y, y_preds))
```
